1. What is a random variable in probability theory?

A random variable is a numerical outcome of a random phenomenon. It assigns a real number to each outcome in a sample space of a random experiment.


---

2. What are the types of random variables?

There are two main types:

Discrete random variables: Take on countable values (e.g., number of heads in 10 coin tosses).

Continuous random variables: Take on values from a continuum or an interval (e.g., height, weight, time).



---

3. What is the difference between discrete and continuous distributions?

Discrete distributions apply to variables that take specific, countable values.

Continuous distributions deal with variables that can take any value within a range.



---

4. What are probability distribution functions (PDF)?

PDFs describe the likelihood of a random variable taking a specific value.

For discrete variables, the PDF gives the probability of each possible outcome.

For continuous variables, the PDF is a curve where the area under the curve in an interval represents the probability.



---

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

A CDF gives the probability that a variable is less than or equal to a value.

A PDF gives the probability of a variable being exactly a value (discrete) or the shape of the probability density (continuous).



---

6. What is a discrete uniform distribution?

A distribution where all outcomes are equally likely. For example, the roll of a fair die has a discrete uniform distribution from 1 to 6.


---

7. What are the key properties of a Bernoulli distribution?

Only two possible outcomes: success (1) or failure (0).

Single trial with probability p of success and 1-p of failure.

Mean = p, Variance = p(1-p).



---

8. What is the binomial distribution, and how is it used in probability?

A distribution of the number of successes in n independent Bernoulli trials.
Used when each trial has two outcomes and a constant probability of success.


---

9. What is the Poisson distribution and where is it applied?

Describes the probability of a given number of events occurring in a fixed interval of time or space.
Common in modeling rare events, like calls received at a call center in an hour.


---

10. What is a continuous uniform distribution?

A distribution where all outcomes in an interval are equally likely.
The probability is uniform across the entire range.


---

11. What are the characteristics of a normal distribution?

Symmetrical, bell-shaped curve

Mean = Median = Mode

Defined by mean (μ) and standard deviation (σ)

Follows the empirical rule (68-95-99.7%)



---

12. What is the standard normal distribution, and why is it important?

A normal distribution with mean 0 and standard deviation 1.
It's used to compute Z-scores and compare values across different normal distributions.


---

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the sampling distribution of the sample mean will be approximately normal, regardless of the population distribution, if the sample size is large enough.
It justifies the use of normal-based inference.


---

14. How does the Central Limit Theorem relate to the normal distribution?

It explains why sample means tend to form a normal distribution, allowing statisticians to apply the normal model in many real-world cases.


---

15. What is the application of Z statistics in hypothesis testing?

Z-statistics are used to determine how far a sample statistic is from the population mean under the null hypothesis.
It helps in calculating p-values and making decisions in hypothesis testing.


---

16. How do you calculate a Z-score, and what does it represent?

Z = (X - μ) / σ
It represents the number of standard deviations a value (X) is from the mean (μ).


---

17. What are point estimates and interval estimates in statistics?

Point estimate: A single value estimate of a parameter (e.g., sample mean).

Interval estimate: A range of values (e.g., confidence interval) believed to contain the parameter.



---

18. What is the significance of confidence intervals in statistical analysis?

Confidence intervals give a range of plausible values for a population parameter and convey the uncertainty of an estimate.


---

19. What is the relationship between a Z-score and a confidence interval?

Z-scores determine the critical values for constructing confidence intervals under the normal distribution.


---

20. How are Z-scores used to compare different distributions?

Z-scores standardize values from different distributions, enabling comparison on a common scale (standard normal).


---

21. What are the assumptions for applying the Central Limit Theorem?

Sample size is large (typically n ≥ 30)

Independent observations

Identically distributed variables with finite mean and variance



---

22. What is the concept of expected value in a probability distribution?

The expected value is the long-run average or mean of a random variable's outcomes, weighted by their probabilities.


---

23. How does a probability distribution relate to the expected outcome of a random variable?

A probability distribution shows all possible values and their probabilities, from which the expected value (mean outcome) can be computed.
